## Purpose of Notebook 

To scrape Google's featured snippets' reference weblink using company names scraped from SGX website. 

Goal is to get all the webpages that directly links to each company's board of directors, then try to scrape these board of directors information from these webpages. 

```python
#importing the required libraries
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests 
import pandas as pd
from lxml import etree
from lxml import html
import sys
```

```python
df = pd.read_excel('scraped_clean.xlsx')
df.shape
```

```python
coy_name_list = df['clean_coy_name'].to_list()
coy_name_list[:5]
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'
}

coyname_1 = []
search_url_1 = []
board_url_1 = []

for i in range(250): 
    # for each company, google search it with keywords Board of Directors
    coyname_1.append(coy_name_list[i])
    query = coy_name_list[i] +"Board of Directors"
    query = query.replace(" ","+")
    url = "https://www.google.com/search?q=" + query
    search_url_1.append(url)
    if requests.get(url,headers=headers,timeout=60).status_code == 200:
        print('now doing... ',i)
        page = requests.get(url,headers=headers,timeout=60)
        tree = html.fromstring(page.content)
        board_url_1.append(tree.xpath('(//div[@class="xpdopen"]//a/@href)[2]'))
    else: #stop the process once there is error
        print('Something went wrong at id', i, 'error code=', requests.get(url,headers=headers,timeout=60).status_code)
        sys.exit()
```
