### Combining both the updated sold out and still selling dataframe

```python
import pandas as pd
from collections import Counter
```

```python
raw = pd.read_csv('cleaned.csv')
so = pd.read_csv('sold_out_cat.csv')
nso = pd.read_csv('not_sold_out.csv')
```

```python
print('raw shape = ' + str(raw.shape))
print('sold out shape = ' + str(so.shape))
print('not sold out shape = ' + str(nso.shape))
```

```python
so.head()
```

```python
# cleaning up the formatting in the sold out dataset
for i in range(len(so)):
    so["Original Price"][i] = so["Original Price"][i].translate(str.maketrans({'[': '', ']': '', "'":""}))
    so['Categories3'][i] = so['Categories3'][i].translate(str.maketrans({'[': '', ']': '', "'":""}))
    so['Brand'][i] = so['Brand'][i].translate(str.maketrans({'[': '', ']': '', "'":""}))
```

```python
so.columns
```

```python
so2 = so[['Index', 'Original Price', 'Categories3', 'Brand']]
so2 = so2.rename(columns = {'Original Price':'Actual_price', 'Categories3':'Catergories','Index':'index'})
so2.head()
```

```python
raw_so = raw[raw['Sold_Out']=='Yes']
raw_so.head()
```

```python
# merging both dataframes together to create sold out dataframe
sold_out = pd.merge(raw_so, so2, on = 'index')
sold_out.columns
```

```python
# creating the first stable version of the dataframe, but categories not spilt yet
merged = pd.concat([sold_out, nso])
merged1 = merged.drop(['Unnamed: 0','index'], axis=1)
merged1.reset_index(inplace=True)
merged1.head()
```

```python
merged1.to_csv('working_df.csv')
```

### Cleaning up the product categories

```python
# cleaned the original categories as there is a lot of duplicate entries
merged1['cleaned_cat'] = ''
for n in range(len(merged1)):
    test = str(merged1['Catergories'][n])
    merged1['cleaned_cat'][n] = ' '.join(dict.fromkeys(test.split()))
```

```python
print('ORIGINAL= ' + merged1['Catergories'][0])
print('CLEANED= ' + merged1['cleaned_cat'][0])
print('//Remove em duplicats//!')
```

```python
main_cat = []
second_cat = []
third_cat = []
for k in merged1.index:
    ca = merged1['cleaned_cat'][k].split(',')[:3]
    main_cat.append(ca[0])
    try:
        second_cat.append(ca[1])
    except:
        second_cat.append('nil')
    try: 
        third_cat.append(ca[2])
    except:
        third_cat.append('nil')
```

```python
indexi = list(merged1['index'])
cat_cleaned = pd.DataFrame(list(zip(indexi, main_cat, second_cat, third_cat)),\
                 columns =['index','main_cat','second_cat','third_cat'])
cat_cleaned.head()
```

```python
final_df = pd.merge(merged1, cat_cleaned, on = 'index')
final = final_df[['Actual_price', 'Brand', 'Catergories', 'Name',
       'Original_price', 'Sale_price', 'Sold_Out', 'Sold_info',
       'Sold_out_time', 'Sold_quantity', 'URL', 'end', 'start',
       'main_cat', 'second_cat', 'third_cat']]
```

```python
# some final data wrangling
final['Brand'] = final['Brand'].replace('no brand','')

for j in final.index:
    if final['second_cat'][j][0] == ' ':
        final['second_cat'][j]=  final['second_cat'][j].replace(final['second_cat'][j],final['second_cat'][j][1:])
    else:
        pass
```

```python
final.head()
```

### Further Data Cleaning for actual price vs original price analysis //not useful anymore as data is wrongly extracted upstream//

```python
## cleaning the actual price as some of them comes with dashes and 2 values
final['Actual_price'] = final['Actual_price'].replace('no actual price','')

### gonna identify those irritating columns, then spilt it, then take average  then add it to a new column
final['Actual_price_clean'] = float(0)
final['Averaged']= ''
for i in final.index:
    if len(final['Actual_price'][i])>9: ##identifying all those price list that comes with a dash and 2 values
        pp = final['Actual_price'][30].replace(' ','').replace('$','').split('-') # splitting into two
        avg = (float(pp[0])+float(pp[1]))/2 #divide and get the average
        final['Actual_price_clean'][i] = avg
        final['Averaged'][i] = 'Averaged'
    else:
        try:
            final['Actual_price_clean'][i] = float(final['Actual_price'][i].replace('$',''))
        except:
            pass 
```

```python
for b in final.index:
    if final['Actual_price_clean'][b] == 0:
        final['Actual_price_clean'][b] = final['Original_price'][b].replace('$ ','')

final['Actual_Price_avg'] = ''
final['Actual_Price_avg'] = '$' + final['Actual_price_clean'].astype(str)
```

```python
final1 = final[['Actual_price', 'Brand', 'Catergories', 'Name', 'Original_price',
       'Sale_price', 'Sold_Out', 'Sold_info', 'Sold_out_time', 'Sold_quantity',
       'URL', 'end', 'start', 'main_cat', 'second_cat', 'third_cat']].rename(columns={'Actual_price':'Actual_price_raw','Catergories':'Categories_raw'})
```

## Discount based on Original Price Data Cleaning

```python
## Cleaning the original price and sale price column, converting it into float columns
for f in final1.index:
    final1['Original_price'][f] = final1['Original_price'][f].replace('$ ','').replace(',','')
    final1['Sale_price'][f] = final1['Sale_price'][f].replace('$ ','').replace(',','')

final1['Original_price'] = final1['Original_price'].astype(float)
final1['Sale_price'] = final1['Sale_price'].astype(float)
```

```python
#creating a discount column
final1['Discount'] = (final1['Original_price'] - final1['Sale_price'])/final1['Original_price']
```

```python
final1.head()
```

```python
three_days = final1[(final1['start'] == 20210908) | (final1['start'] == 20210909) | (final1['start'] == 20210910)]
```

```python
gb = three_days.groupby(['main_cat','start'])['Discount'].mean().reset_index()
gb.head()
```

```python
gb_pivot = gb.pivot(index='main_cat', columns='start').rename_axis(None)
gb_pivot.columns = gb_pivot.columns.droplevel() # remove multilevel indexing caused by pivot table
gb_pivot = gb_pivot.reset_index()
gb_pivot.rename(columns = {'index':'main_cat'}, inplace=True)
gb_pivot.head()
```

```python
gb_pivot.to_excel('discount_cat_date.xlsx')
```
